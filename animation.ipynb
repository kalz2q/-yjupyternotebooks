# color circle

```
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

nx = 20
ny = 20

fig = plt.figure()
plt.axis([0,nx,0,ny])
ax = plt.gca()
ax.set_aspect(1)

def init():
    # initialize an empty list of cirlces
    return []

def animate(i):
    # draw circles, select to color for the circles based on the input argument i. 
    someColors = ['r', 'b', 'g', 'm', 'y']
    patches = []
    for x in range(0,nx):
        for y in range(0,ny):
            patches.append(ax.add_patch( plt.Circle((x+0.5,y+0.5),0.45,color=someColors[i % 5]) ))
    return patches

anim = FuncAnimation(fig, animate, init_func=init, frames=10, interval=200, blit=True)
plt.close()
from IPython.display import HTML
HTML(anim.to_jshtml())
```
